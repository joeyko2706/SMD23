```python
import numpy as np
import scipy.constants as const
import scipy as sc
import matplotlib.pyplot as plt
```

### $f(v)$ as a function of $v_m$

```python
def ff(v, m, T):
    return (m/(2*np.pi*const.Boltzmann*T))**(3/2) * np.exp(-(m*v**2)/(2*const.Boltzmann*T))*4*np.pi*v**2


def f(v,v_m):
    return (1/v_m)**3 *np.exp(-(v/v_m)**2)*4*np.pi*v**2
```

```python
v = np.linspace(0,4,1000)


plt.plot(v, f(v, 1),"purple")
plt.tight_layout(pad=0, h_pad=1.08, w_pad=1.08)
plt.xlim(0,4)
plt.xlabel("velocity $v/v_m$")
plt.ylabel("velocity distribution")

plt.show()
```

### calculating the means

```python
v_m = np.linspace(0.1,6.1,100)
mean = np.zeros(len(v_m))
n=0
for i in v_m:
    mean[n] = np.mean(f(v, i))
    n=n+1

plt.plot(v_m, mean,"purple")
plt.tight_layout(pad=0, h_pad=1.08, w_pad=1.08)
plt.xlim(0.1,6.1)
plt.xlabel("$v_m$")
plt.ylabel("mean")

plt.show()
```
