## (b)

```python
# import the packages needed
import numpy as np
import scipy.constants as const
import scipy as sc
import matplotlib.pyplot as plt
```

```python
# define all the constants

k_B = const.Boltzmann
m = const.physical_constants["molar mass constant"]
m = m[0]*m[2] # implementing the molecular mass as a float64, instead of a tuple
temperature = np.array([100, 300, 500, 1000, 5000])#.astype('float') #save as float64
N = (m/(2*np.pi*k_B*temperature))**(3/2)
v_m = np.sqrt((2*k_B*temperature)/m)
#print(v_m) #print the the most probable velocities to 

v = np.linspace(0,0.05,1000)

for temp in temperature: # define the velocity dist. and plot it
    N = (m/(2*np.pi*k_B*temp))**(3/2)
    f_temp = N * np.exp(-(m*v**2)/(2*k_B*temp))*4*np.pi*v**2
    plt.plot(v, f_temp, label='T={} K'.format(temp))
    plt.legend()
    plt.tight_layout(pad=0, h_pad=1.08, w_pad=1.08)
    plt.xlim(0,0.05)
    plt.xlabel(r'velocity $v$ [m/s]')
    plt.ylabel(r'velocity distribution')

plt.show()
```

```python
# calculate the mean values of each velocity distribution

for temp in temperature:
#    print(f_temp)
    print(r'<v_{}> = '.format(temp), np.mean(f_temp))
```

## (c)

```python
from scipy.signal import find_peaks, peak_widths 

# calculate the median of each velocity distribution

for temp in temperature:
    peaks, properties = find_peaks(f_temp)
    print(r'v_{} = '.format(temp), f_temp[peaks])
```

## (d)

```python
for temp in temperature:
    results_half = peak_widths(f_temp, peaks, rel_height=0.5)
    print(results_half[0])  # widths
```
