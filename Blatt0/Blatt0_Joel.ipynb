```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('height_weight.csv')
#df.head()
#df.tail()

#plt.subplot(3, 2)
#plt.hist(df.height, df.weight, bins = '5')

# Histogramm Optionen

hist_options = dict(
    histtype='step',
    lw=2,
)

height = df['height']
weight = df['weight']


fig = plt.figure()
plt.subplot(3, 2, 1)
plt.hist(height, bins = 5, label='5 Bins, height')
plt.xlabel('Height')
plt.legend(loc='best')


plt.subplot(3, 2, 2)
plt.hist(height, bins = 10, label='10 Bins, height')
plt.xlabel('Height')
plt.legend(loc='best')


plt.subplot(3, 2, 3)
plt.hist(height, bins = 15, label='15 Bins, height')
plt.xlabel('Height')
plt.legend(loc='best')


plt.subplot(3, 2, 4)
plt.hist(height, bins = 20, label='20 Bins, height')
plt.xlabel('Height')
plt.legend(loc='best')


plt.subplot(3, 2, 5)
plt.hist(height, bins = 30, label='30 Bins, height')
plt.xlabel('Height')
plt.legend(loc='best')

plt.subplot(3, 2, 6)
plt.hist(height, bins = 50, label='50 Bins, height')
plt.xlabel('Height')
plt.legend(loc='best')
plt.show()
```

```python
fig = plt.figure()
plt.subplot(3, 2, 1)
plt.hist(weight, bins = 5, label='5 Bins, weight')
plt.xlabel('Weight')
plt.legend(loc='best')


plt.subplot(3, 2, 2)
plt.hist(weight, bins = 10, label='10 Bins, weight')
plt.xlabel('Weight')
plt.legend(loc='best')


plt.subplot(3, 2, 3)
plt.hist(weight, bins = 15, label='15 Bins, weight')
plt.xlabel('Weight')
plt.legend(loc='best')


plt.subplot(3, 2, 4)
plt.hist(weight, bins = 20, label='20 Bins, weight')
plt.xlabel('Weight')
plt.legend(loc='best')


plt.subplot(3, 2, 5)
plt.hist(weight, bins = 30, label='30 Bins, weight')
plt.xlabel('Weight')
plt.legend(loc='best')

plt.subplot(3, 2, 6)
plt.hist(weight, bins = 50, label='50 Bins, weight')
plt.xlabel('Weight')
plt.legend(loc='best')
plt.show()
```

For both the histogram showing the height and the weight it seems to be enough to plot 30 bins as there are blank spaces in the histograms with 50 bins each.

Using a dataset with more than 250 people will most likely show a gaussian distribution, if the bins are set correctly.

A reasonable bin with would be one that bins the data in a way that only one data-point is shown. So for the height use 1/2 for 1 cm and for the weight use 1 for 1 kg.

```python
dist = np.linspace(1,100, 10**5)
binning5 = np.geomspace(1,100, 5)
binning10 = np.geomspace(1,100, 10)
binning15 = np.geomspace(1,100, 15)
binning20 = np.geomspace(1,100, 20)
binning30 = np.geomspace(1,100, 30)
binning50 = np.geomspace(1,100, 50)

fig = plt.figure()
plt.subplot(3,2, 1)
plt.hist(dist, binning5)
plt.xscale('log')

plt.subplot(3, 2, 2)
plt.hist(dist, binning10)
plt.xscale('log')

plt.subplot(3, 2, 3)
plt.hist(dist, binning15)
plt.xscale('log')

plt.subplot(3, 2, 4)
plt.hist(dist, binning20)
plt.xscale('log')

plt.subplot(3, 2, 5)
plt.hist(dist, binning30)
plt.xscale('log')

plt.subplot(3, 2, 6)
plt.hist(dist, binning50)
plt.xscale('log')
```

When using more bins the strucure of the data is more clearly visible.
